```
from bokeh.sampledata.iris import flowers as ir
ir.head()
```

```
from bokeh.palettes import Category20
from bokeh.transform import factor_cmap

source = ColumnDataSource(ir)
cc = factor_cmap('species',['red','blue','yellow'], list(set(ir.species.to_list())))

p = figure()
p.circle('sepal_length', 'sepal_width', color=cc, source = source)
show(p)
```

```
from bokeh.layouts import gridplot
from bokeh.models import HoverTool

pset = {'plot_height':400, 'plot_width':400, 'tools':'pan,wheel_zoom,box_select'}

x = figure(**pset)
x.circle('sepal_length', 'sepal_width', color=cc, source = source)

y = figure(x_range=x.x_range, y_range=x.y_range,**pset)
y.square('petal_length', 'petal_width', color=cc, source = source)


p = gridplot([[x,y]])
show(p)
```

```
from bokeh.sampledata.iris import flowers as ir
from bokeh.models.widgets import Select, Button
from bokeh.plotting import output_notebook, show, figure
from bokeh.models import ColumnDataSource, CustomJS, Select
from bokeh.transform import factor_cmap
from bokeh.layouts import column
from bokeh import events



output_notebook()
```

```
source = ColumnDataSource(ir)
cc = factor_cmap('species', ['firebrick', 'navy', 'darkorange'], list(set(ir.species)))

# selector = Select(title="Option:", value='setosa', options = list(set(ir.species)))
btn = Button(label='Change', button_type='warning')

change = CustomJS(args={'btn':btn, 'source':source}, code="""
  source.data.petal_length = source.data.petal_length.map(x=>0);
  source.change.emit();

""")

p = figure(plot_height=400, plot_width=400)
p.circle(x='petal_length', y='petal_width', color=cc, source=source)
btn.js_on_event(events.ButtonClick, change)


show(column(btn,p))
```
