# Bar Chart

```
from pyecharts.charts import Bar
from pyecharts import options as opts
from pyecharts.globals import ThemeType

bar = (Bar(init_opts=opts.InitOpts(theme=ThemeType.LIGHT))
  .add_xaxis(["Shirt", "Pants", "Suit", "Shoe", "Socks", "Jumper"])
  .add_yaxis("StoreA", [5, 20, 36, 10, 75, 90])
  .add_yaxis("StoreB", [4, 56, 22, 52, 77, 30])
  )
bar.render('abc.html')

from IPython import display

display.HTML('abc.html')
```

# Line Chart

```
import pyecharts.options as opts
from pyecharts.charts import Line
import numpy as np


x_data = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
y_data = [820, 932, 901, 934, 1290, 1330, 1320]
y_data2 = np.random.randint(600,2000,7)
line = (Line()
  .add_xaxis(x_data)
  .add_yaxis('customers', y_data)
  .add_yaxis('customers2', y_data2.tolist())
  .render('line.html')
)
display.HTML('line.html')
```

# Map

```
from pyecharts import options as opts
from pyecharts.charts import Map
from pyecharts.faker import Faker

c = (
    Map()
    .add("ScoreA", [list(z) for z in zip(Faker.country, Faker.values())], "world")
    .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
    .set_global_opts(
        title_opts=opts.TitleOpts(title="Map-world"),
        visualmap_opts=opts.VisualMapOpts(max_=200),
    )
    .render("map_world.html")
)
display.HTML('map_world.html')
```

```
from pyecharts import options as opts
from pyecharts.charts import Bar, Grid, Line
from pyecharts.faker import Faker

bar = (
    Bar()
    .add_xaxis(Faker.choose())
    .add_yaxis("商家A", Faker.values())
    .add_yaxis("商家B", Faker.values())
    .set_global_opts(title_opts=opts.TitleOpts(title="Grid-Bar"))
)
line = (
    Line()
    .add_xaxis(Faker.choose())
    .add_yaxis("商家A", Faker.values())
    .add_yaxis("商家B", Faker.values())
    .set_global_opts(
        title_opts=opts.TitleOpts(title="Grid-Line", pos_top="48%"),
        legend_opts=opts.LegendOpts(pos_top="48%"),
    )
)

grid = (
    Grid()
    .add(bar, grid_opts=opts.GridOpts(pos_bottom="60%"))
    .add(line, grid_opts=opts.GridOpts(pos_top="60%"))
    .render("grid_vertical.html")
)
display.HTML("grid_vertical.html")
```
